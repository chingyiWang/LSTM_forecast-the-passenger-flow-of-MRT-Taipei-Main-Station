```python
# 北車，共1個
# 2017.2018年

import pandas as pd
import matplotlib.pyplot as plt

df_r11_2017 = pd.read_csv('C:/Users/ching/Desktop/final project/data/MRT_Data_complete_0819/MRT_Data_2017_complete_0819/台北車站.csv')
df_r11_2018 = pd.read_csv('C:/Users/ching/Desktop/final project/data/MRT_Data_complete_0819/MRT_Data_2018_complete_0819/台北車站.csv')

#C:/Users/ching/Desktop/final project/data/MRT_Data_complete_0819/MRT_Data_2017_complete_0819
```

```python
cluster6_1718 = pd.concat([df_r11_2017,df_r11_2018])
cluster6_1718.shape
```

```python
cluster6_1718['Sum_people'] = cluster6_1718['Get_In_Count']+cluster6_1718['Way_Out_Count']
cluster6_1718.head()
```

```python
#設儲存路徑及檔名
Result ='C:/Users/ching/Desktop/final project/data/MRT_Data_complete_0819/8cluster_6/8cluster_6_train.csv'

#to.csv
cluster6_1718.to_csv(Result, index=True )

print( '成功產出'+ Result)
```
